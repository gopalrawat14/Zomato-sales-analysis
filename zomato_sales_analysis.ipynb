# STEP:2 create the dataframe





```python
dataframe= pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# STEP:3 cleaning the data

```python
#for that converting the data type of the column rating
def handleRate(value):
  value=str(value).split('/')
  value=value[0]
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# QUESTION 1: What type of restaurant do the majority of customers order from?
# for that we will show it using graphs, using seaborn library

```python
dataframe.head()

#it shows only top 5 values of the graph
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurent")
```

# Question 2: How many votes has each type of restaurant received from customers?

```python
dataframe.head()
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})  # Corrected 'DataFrame'
plt.plot(result, c='blue', marker='o')
plt.xlabel("Type of restaurant", c="red", size=15)
plt.ylabel("Vote", c='red', size=15)
```

# Conclusion: Dining restaurents has recieved the maximum votes.

# Question 5: Which mode (online or offline) has received the maximum rating?

```python
dataframe.head()
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(x= "online_order", y="rate", data= dataframe)
```

# Conclusion: Offline order received lower rating in comparison to online order.

# Question 6: Which type of restaurant received more offline orders, so that Zomato can suggest customers with some good offers?

```python
dataframe.head()
```

```python
pivot_table = dataframe.pivot_table (index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed in (Type)")
plt.show()
```

# Conclusion: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer orders in person at restaurants, but prefer online ordering at cafes.
